```python
import array as arr

nilai = arr.array('i', [80, 85, 90, 75])
print("Array awal:", nilai)

nilai.append(88)
print("Setelah append 88:", nilai)

nilai.insert(1, 95)
print("Setelah insert 95 di indeks 1:", nilai)
```

```python
import array as arr

nilai = arr.array('i', [80, 85, 90, 75])

# print("Array:", nilai)
print("Elemen pertama:", nilai[0])
print("Elemen keempat:", nilai[3])
```

```python
import array as arr

nilai = arr.array('i', [80, 85, 90, 75, 88])
print("Array awal:", nilai)

nilai.remove(85)
print("Setelah remove 85:", nilai)

nilai.pop()
print("Setelah pop():", nilai)
```

```python
import array as arr

nilai = arr.array('i', [80, 85, 90, 75, 88])
print("Array awal:", nilai)

print("Slicing indeks 1 sampai 4:", nilai[1:4])
print("Dari awal sampai indeks 3:", nilai[:3])
print("Dibalik:", nilai[::-1])
```

```python
import array as arr

nilai = arr.array('i', [80, 85, 90, 75])
print("Array awal:", nilai)

nilai[0] = 100
nilai[2] = 95

print("Setelah perubahan:", nilai)
```

```python
import array as arr

nilai = arr.array('i', [80, 90, 75, 90])
print("Array awal:", nilai)

print("Jumlah angka 90:", nilai.count(90))

nilai.reverse()
print("Setelah reverse:", nilai)

nilai.extend([70, 60])
print("Setelah extend:", nilai)
```
